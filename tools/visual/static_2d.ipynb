```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
import numpy as np
np.random.seed(114514)
y = np.random.standard_normal(20)
x = np.arange(len(y)) # 固定的整数值
plt.plot(y)
```

```python
plt.plot(y.cumsum())
plt.grid(False)
plt.axis('image')
```

```python
plt.plot(y.cumsum())
plt.xlim(0, 20)
plt.ylim(np.min(y.cumsum()) - 1, np.max(y.cumsum()) + 1)
```

```python
plt.figure(figsize = (10, 6)) # 增大尺寸
plt.plot(y.cumsum(), 'black', lw = 1.5) # 线宽为1.5的蓝色线
plt.plot(y.cumsum(), 'gs') # 加上红点
plt.xlabel('index') # x轴加标签
plt.ylabel('index') # y轴加标签
plt.title('Plot Demo')
```
