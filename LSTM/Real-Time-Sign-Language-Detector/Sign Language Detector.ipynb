### use MediaPipe's Holistic models to track keypoints from images

```python
# Getting the holistic model and initializing it
mp_holistic = mp.solutions.holistic
# Getting the drawing utilities for plotting points
mp_drawing = mp.solutions.drawing_utils
```

```python
# Function to detect points in our image
def mp_detection(image, model):
    # Converting BGR image of OpenCV into RGB
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    rgb_image.flags.writeable = False # Non-writeable
    
    # Process the image using mediapipe model
    result = model.process(rgb_image)
    rgb_image.flags.writeable = True # Writeable
    
    # Changing it back to BGR format for OpenCV
    bgr_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2BGR)
    return bgr_image, result
```

### We need to access our camera as we loop through all frames and render them 

```python
# Plotting all keypoints on the image
def draw_landmarks(image, result):
    mp_drawing.draw_landmarks(image, result.face_landmarks, mp_holistic.FACEMESH_CONTOURS)
    mp_drawing.draw_landmarks(image, result.pose_landmarks, mp_holistic.POSE_CONNECTIONS)
    mp_drawing.draw_landmarks(image, result.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
    mp_drawing.draw_landmarks(image, result.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
```

```python
# Plotting all keypoints on the image
def draw_styled_landmarks(image, result):
    # Draw face connections
    mp_drawing.draw_landmarks(image, result.face_landmarks, mp_holistic.FACEMESH_CONTOURS,
                             mp_drawing.DrawingSpec(color=(80, 110, 10), thickness=1, circle_radius=1),
                             mp_drawing.DrawingSpec(color=(80, 256, 121), thickness=1, circle_radius=1)
                             )
    # Draw pose connections
    mp_drawing.draw_landmarks(image, result.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2)
                             ) 
    # Draw left hand connections
    mp_drawing.draw_landmarks(image, result.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                             mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
                             ) 
    # Draw right hand connections  
    mp_drawing.draw_landmarks(image, result.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                             mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                             ) 
```

```python
cam = cv2.VideoCapture(0)
# Accessing MP Model
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cam.isOpened():
        # Reading current frame from video
        retval, frame = cam.read()

        # Detect points/landmarks and draw them
        image, result = mp_detection(frame, holistic)
        draw_styled_landmarks(image, result)

        # Display frame
        cv2.imshow('Rendered Output', image)
        
        # Break if we press 'q'
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    # Stop capturing video 
    cam.release()
    cv2.destroyAllWindows()
```

```python
def extract_keypoints(result):
    # Each pose landmark has 4 attributes so total values = 33 x 4
    pose = np.array([[res.x, res.y, res.z, res.visibility] for res in result.pose_landmarks.landmark]).flatten() if result.pose_landmarks else np.zeros(33*4)
    lh = np.array([[res.x, res.y, res.z] for res in result.left_hand_landmarks.landmark]).flatten() if result.left_hand_landmarks else np.zeros(21*3)
    rh = np.array([[res.x, res.y, res.z] for res in result.right_hand_landmarks.landmark]).flatten() if result.right_hand_landmarks else np.zeros(21*3)
    face = np.array([[res.x, res.y, res.z] for res in result.face_landmarks.landmark]).flatten() if result.face_landmarks else np.zeros(468*3)
    return np.concatenate([pose, face, lh, rh])
```

### Now we will setup our directories for storing our extracted keypoints

```python
DATA_PATH = os.path.join('MP_Data')

# Actions that we try to detect
actions = np.array(['Hello', 'Thanks', 'I Love You'])
# Data consisting of 30 videos/sequences each of length = 30 frames
no_sequences = 30
sequence_length = 30
```

```python
for action in actions:
    for sequence in range(no_sequences):
        try:
            os.makedirs(os.path.join(DATA_PATH, action, str(sequence)))
        except:
            pass
```

### Now we need to gather data for training

```python
cam = cv2.VideoCapture(0)
# Accessing MP Model
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    for action in actions:
        for sequence in range(no_sequences):
            for frame_count in range(sequence_length):
                # Reading current frame from video
                retval, frame = cam.read()

                # Detect points/landmarks and draw them
                image, result = mp_detection(frame, holistic)
                draw_styled_landmarks(image, result)
                
                # Setting durations and counters
                if frame_count == 0:
                    cv2.putText(image, 'Starting Collection...', (120, 500),
                               cv2.FONT_HERSHEY_DUPLEX, 1, (0, 255, 0), 4, cv2.LINE_AA)
                    cv2.putText(image, 'Collecting frames for {} for Video {}'.format(action, sequence), (15,12),
                                cv2.FONT_HERSHEY_DUPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)
                    # Display frame
                    cv2.imshow('Rendered Output', image)

                    cv2.waitKey(1500)
                else:
                    cv2.putText(image, 'Collecting frames for {} for Video {}'.format(action, sequence), (15,12),
                                cv2.FONT_HERSHEY_DUPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)
                    # Display frame
                    cv2.imshow('Rendered Output', image)

                # Extracting and exporting keypoints from the frame
                keypoints = extract_keypoints(result)
                npy_path = os.path.join(DATA_PATH, action, str(sequence), str(frame_count))
                np.save(npy_path, keypoints)
                
                
                # Break if we press 'q'
                if cv2.waitKey(10) & 0xFF == ord('q'):
                    break
    # Stop capturing video 
    cam.release()
    cv2.destroyAllWindows()
```

### Now we need to preprocess our data 

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
label_map = {label:num for num, label in enumerate(actions)}
label_map
```

```python
sequences, labels = [], []
for action in actions:
    for sequence in range(no_sequences):
        window = []
        for frame_count in range(sequence_length):
            res = np.load(os.path.join(DATA_PATH, action, str(sequence), "{}.npy".format(frame_count)))
            window.append(res)
        sequences.append(window)
        labels.append(label_map[action])
```

```python
X = np.array(sequences)
X.shape
```

```python
# Creates one hot-encoded label array
y = to_categorical(labels).astype(int)
y.shape
```

```python
# Splitting data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Now we will build our model for prediction

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import TensorBoard
```

```python
from tensorflow.python.client import device_lib
device_lib.list_local_devices()
```

```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '/device:CPU:0'
```

```python
log_dir = os.path.join('logs')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
model = Sequential()
```

```python
model.add(LSTM(64, return_sequences=True, activation='relu', input_shape=(30, 1662)))
model.add(LSTM(128, return_sequences=True, activation='relu'))
model.add(LSTM(64, return_sequences=False, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(actions.shape[0], activation='softmax'))
```

```python
model.summary()
```

```python
# Model compilation
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
# Model training
model.fit(X_train, y_train, epochs=150, callbacks=[tb_callback])
```

### Let us check the score of our model

```python
def calculate_score(model, X_test, y_test):
    res = model.predict(X_test)
    total = X_test.shape[0]
    count = 0
    for i in range(total):
        if actions[np.argmax(res[i])] == actions[np.argmax(y_test[i])]:
            count = count+1
    print("Model Prediction Accuracy: ", (count*100)/total, '%')
    return
```

```python
calculate_score(model, X_test, y_test)
```
